#5) Membersihkan data

```python
#Mengubah tipe data
#data_drop['Diagnosis']=data_drop['Diagnosis'].astype(str).astype(float)

from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
data_drop['Diagnosis']=le.fit_transform(data_drop['Diagnosis'])
#for col in cat_cols:
# data_drop[col]=le.fit_transform(data_drop[col])
```

```python
data_drop.info()
```

```python
data_drop.duplicated().sum()
```

```python
data_drop.isnull().sum()
```

```python
#Menampilkan data outlayer
plt.figure(figsize=(20,20))
sns.boxplot(data =data_drop,palette='Set2',orient='h')
plt.title(" ",fontsize=10)
plt.show()
```

```python
#Melakukan normalisasi outlayer dengan menggunakan metode IQR
data_iqr=data_drop.copy()
for col in data_iqr.select_dtypes(include=['int64','float64']):
  if col not in['Diagnosis','concave_points3']:
    Q1=data_drop[col].quantile(0.25)
    Q3=data_drop[col].quantile(0.75)
    IQR=Q1-Q3
    lower_bound=Q1-1.5 * IQR
    upper_bound=Q3+1.5 *IQR

    data_iqr[col]=data_drop[col].apply(lambda x:lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
    print(f" outlier pada'{col}' telah ditangani dengan IQR")
```

```python
plt.figure(figsize=(20,20))
sns.boxplot(data=data_iqr, orient='h',palette='Set2')
plt.title(" ",fontsize=10)
plt.show()
```

#7) Nenentukan Label

```python
X=data_iqr.drop('Diagnosis', axis=1)
y=data_iqr['Diagnosis']
```

```python
data_drop['Diagnosis'].value_counts().plot(kind='bar',figsize=(8,5),color=['green','Blue'] )
plt.title("Target Kelas Diagnosis",fontsize=10)
plt.show()
```

```python
from imblearn.over_sampling  import SMOTE
from sklearn.model_selection import train_test_split
```

```python
smote=SMOTE(random_state=42)
x_smote,y_smote=smote.fit_resample(X,y)
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x=y_smote)
plt.show()
```

```python
#pembagian data dengan jumlah data uji 20 % dan data train 80%
X_train,X_test,y_train,y_test= train_test_split(x_smote,y_smote, test_size=0.2,random_state=42,stratify=y_smote)
```

#8) Membangun Model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
```

```python
dt= DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
lr= LogisticRegression()
lr.fit(X_train,y_train)
```

```python
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
nb=GaussianNB()
nb.fit(X_train,y_train)
```

#9) Mengevaluasi Model

```python
y_pred_dt=dt.predict(X_test)
print("Accuracy: ",accuracy_score(y_test,y_pred_dt))
print("\nClassification report Decision TREE: ")
print(classification_report(y_test,y_pred_dt))
```

```python
cm_dt=confusion_matrix(y_test,y_pred_dt)
sns.heatmap(cm_dt,annot=True, fmt="d", cmap="Blues",
            yticklabels=['M','B'],
            xticklabels=['M','B'])
plt.title("Confussion Matriks dari Decision Tree")
plt.xlabel("Predict")
plt.ylabel("Actual")
plt.show()
```

```python
y_pred_lr=lr.predict(X_test)
print("Accuracy Score: ",accuracy_score(y_test,y_pred_lr))
print("\nLaporan klasifikasi dari Model Logistik Regression")
print(classification_report(y_test,y_pred_lr))
```

```python
cm_lr=confusion_matrix(y_pred_lr,y_test)
sns.heatmap(cm_lr, annot=True,fmt='d', cmap="Blues",
            xticklabels=['M','B'],
            yticklabels=['M','B'])
plt.title(" Confusion Matrik dari Logistik Regression")
plt.xlabel("Predict")
plt.ylabel("Actual")
plt.show()
```

```python
y_pred_rf=rf.predict(X_test)
print("Accuracy: ", accuracy_score(y_pred_rf,y_test))
print("Lporan Klasifikasi Dari model Random FOREST")
print(classification_report(y_pred_rf,y_test))
```

```python
cm_rf=confusion_matrix(y_pred_rf,y_test)
sns.heatmap(cm_rf, annot=True, fmt='d', cmap="Blues",
            xticklabels=['M','B'],
            yticklabels=['M','B'])
plt.xlabel('Predict')
plt.ylabel('Actual')
plt.title("Confussion Matrix dari Random Forest")
plt.show()
```

```python
y_pred_nb=nb.predict(X_test)
print("Accuracy score : ", accuracy_score(y_test,y_pred_nb))
print("\n Laporan Klasifikasi Naive Bayes")
print(classification_report(y_pred_nb,y_test))
```

```python
cm_nb=confusion_matrix(y_pred_nb,y_test)
sns.heatmap(cm_nb, annot=True,fmt='d', cmap="Blues",xticklabels=['M','B'], yticklabels=['M','B'])
plt.xlabel("Predict")
plt.ylabel("Actual")
plt.title("Confussion Matriks dari Naive Bayes")
plt.show()
```

```python
model_names=['Decision Tree','Logistic Regression','Random Forest','Naive Bayes']
accuracies=[
    accuracy_score(y_test,y_pred_dt),
    accuracy_score(y_test,y_pred_lr),
    accuracy_score(y_test,y_pred_rf),
    accuracy_score(y_test,y_pred_nb)
]
```

```python
plt.figure(figsize=(8,5))
plt.bar(model_names,accuracies, color=['green','Blue'])
plt.title("Perbandingan Akurasi dari Setiap Model")
plt.xlabel("Model yang digunakan")
plt.ylabel("Akurasi dari model")
plt.xticks(rotation=0,ha='right')
plt.ylim(0,1)

for i,v in enumerate(accuracies):
  plt.text(i,v+0.01,f'{v:2f}',ha='center',va='bottom')

plt.tight_layout()
plt.show()
```
